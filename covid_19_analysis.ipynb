```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("country_wise_latest.csv")
df.info()
```

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df=df.drop(columns=["Deaths / 100 Recovered"])
df
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="WHO Region", y=np.log1p(df["Confirmed"]), data=df)
plt.title("Distribution of Confirmed Cases by WHO Region")
plt.ylabel("log(Confirmed +1)")
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries=df.nlargest(10, "Confirmed")

plt.figure(figsize=(12,6))
sns.barplot(x="Confirmed", y="Country/Region", data=top_countries, palette="viridis")
plt.xlabel("Number of Confirmed Cases")
plt.ylabel("Country/Region")
plt.title("Top 10 Countries by affected by COVID-19 Cases")
plt.show()
```

```python
top10_deaths=df.sort_values("Deaths / 100 Cases", ascending=False).groupby("WHO Region").head(10)
regions=top10_deaths["WHO Region"].unique()
num_regions=len(regions)
fig, axes=plt.subplots(num_regions, 1, figsize=(12, num_regions*4))
for i, region in enumerate(regions):
    data=top10_deaths[top10_deaths["WHO Region"]==region]
    sns.barplot(x="Deaths / 100 Cases", y="Country/Region", data=data, ax=axes[i], palette="vanimo")
    axes[i].set_title(f"Top 10 Countries by Death Rate in {region}")
    axes[i].set_xlabel("Death Rate (%)")
    axes[i].set_ylabel("Country/Region")
    plt.tight_layout()
plt.show()
```

top 10 countries for the recovery rate in each region

```python
top10_recovered=(df.sort_values("Recovered / 100 Cases", ascending=False).groupby("WHO Region").head(10))
regions=top10_recovered["WHO Region"].unique()
num_regions=len(regions)

fig, axes=plt.subplots(num_regions, 1, figsize=(10, num_regions*4))

for i, region in enumerate(regions):
    data=top10_recovered[top10_recovered["WHO Region"]==region]
    sns.barplot(x="Recovered / 100 Cases", y="Country/Region", data=data, ax=axes[i], palette="vanimo")
    axes[i].set_title(f"Top 10 Countries by Recovery Rate in {region}")
    axes[i].set_xlabel("Recovery Rate (%)")
    axes[i].set_ylabel("Country/Region")
    plt.tight_layout()
```

top 10 countries with fastest spread in each region

```python
top10_spread_by_region=df.sort_values("1 week % increase", ascending=False).groupby("WHO Region").head(10)
regions=top10_spread_by_region["WHO Region"].unique()
num_regions=len(regions)
fig, axes=plt.subplots(num_regions, 1, figsize=(10, num_regions*4))
for i, region in enumerate(regions):
    data=top10_spread_by_region[top10_spread_by_region["WHO Region"]==region]
    sns.barplot(x="1 week % increase", y="Country/Region", data=data, ax=axes[i], palette="vanimo")
    axes[i].set_title(f"Top 10 Countries by Spread Rate in {region}")
    axes[i].set_xlabel("Spread Rate (%)")
    axes[i].set_ylabel("Country/Region")
    plt.tight_layout()
plt.show()
```

relationship between death and recovery rate

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Deaths / 100 Cases", y="Recovered / 100 Cases", hue="WHO Region", data=df, palette="magma",alpha=0.7   )
plt.title("Death Rate vs Recovery Rate by WHO Region")
plt.xlabel("Death Rate (%)")
plt.ylabel("Recovery Rate (%)")
plt.legend(title="WHO Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
top10_confirmed=df.nlargest(10, "Confirmed")
plt.figure(figsize=(12,6))
sns.scatterplot(x="Deaths / 100 Cases", y="Recovered / 100 Cases", hue="Country/Region", data=top10_confirmed, palette="tab10"
                , alpha=0.8, sizes=(100,1000), size="Confirmed")
plt.title("Death Rate vs Recovery Rate for Top 10 Countries by Confirmed Cases")
plt.xlabel("Death Rate (%)")
plt.ylabel("Recovery Rate (%)")
plt.legend(title="Country/Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
